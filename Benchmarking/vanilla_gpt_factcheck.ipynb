```python
def classify_video(transcription, categories):
  chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": f"You are a model that takes in a transcription of a video and outputs a number prediction pertaining to whether the video is misinforming or not misinforming. Classify the following transcription of a short-form video: '{transcription}' into one of the following categories: '{categories}'. Only output a single number (your prediction); do not have words in your response.",
        }
    ],
    model="gpt-3.5-turbo",
    max_tokens = 1,
  )

  return chat_completion;

categories_list = ["0 = true; 1 = mostly true; 2 = half true; 3 = mostly false; 4 = false; 5 = pants on fire", "0 = not misinforming; 1 = misinforming"]
```

```python
transcription = "Abraham Lincoln was never shot. He is alive!"
output1 = classify_video(transcription, categories_list[0]).choices[0].message.content
output2 = classify_video(transcription, categories_list[1]).choices[0].message.content

print(f"0-5:{output1}, 0-1:{output2}")
```
